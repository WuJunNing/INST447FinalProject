```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
semester = '202301'
dept = 'INST'
# https://app.testudo.umd.edu/soc/search?courseId=INST&sectionId=&termId=202301&_openSectionsOnly=on&creditCompare=%3E%3D&credits=0.0&courseLevelFilter=UGRAD&instructor=&_facetoface=on&_blended=on&_online=on&courseStartCompare=&courseStartHour=&courseStartMin=&courseStartAM=&courseEndHour=&courseEndMin=&courseEndAM=&teachingCenter=ALL&_classDay1=on&_classDay2=on&_classDay3=on&_classDay4=on&_classDay5=on
url = f"https://app.testudo.umd.edu/soc/search"
params = {'courseId': dept,
          'termId': semester,
          'courseStartCompare': '',
          'courseStartHour': '',
          'courseStartMin': '',
          'courseStartAM': '',
          'courseLevelFilter': 'UGRAD'}
response = requests.get(url, params=params)
response.url
```

```python
soup = BeautifulSoup(response.text)
soup.title.string
```

```python
soup.find_all('div', 'course')
```

```python
containers = soup.find_all('div', 'course')
courses_list = []
for crs in containers:
    course_id = crs.find('div', 'course-id').text
    course_title = crs.select_one('.course-title').text
    section_info = []
    for sections in crs.select('.section-info-container'):
        section_info.append({'section_id': sections.select_one('.section-id').text,
                         'instructor': sections.select_one('.section-instructors').text})
    courses_list.append({'course_id': course_id,
                         'course_title': course_title,
                         'sections': section_info})

pd.DataFrame(courses_list)
```
